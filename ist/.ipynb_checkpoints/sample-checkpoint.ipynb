```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from yahoofinancials import YahooFinancials
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression


assets = ['TSLA', 'MSFT', 'FB']
yahoo_financials = YahooFinancials(assets)
scaler = MinMaxScaler(feature_range=(0, 1))

data = yahoo_financials.get_historical_price_data(start_date='2020-01-01',
                                                  end_date='2020-12-31',
                                                  time_interval='daily')

df = pd.DataFrame(data['TSLA']['prices'])
df.drop('date', axis=1,inplace=True)
df['volume'] = df['volume'].astype(float)
df['formatted_date'] = pd.to_datetime(df.formatted_date,format='%Y-%m-%d')
df.index = df['formatted_date']

#sorting
data = df.sort_index(ascending=True, axis=0)

#creating a separate dataset
new_data = pd.DataFrame(index=range(0,len(df)),columns=['formatted_date', 'adjclose'])

for i in range(0,len(data)):
    new_data['formatted_date'][i] = data['formatted_date'][i]
    new_data['adjclose'][i] = data['adjclose'][i]
    
    #create features
from fastai.tabular import  add_datepart
add_datepart(new_data, 'formatted_date')
new_data.drop('Elapsed', axis=1, inplace=True)  #elapsed will be the time stamp

#split into train and validation
train = new_data[:200]
valid = new_data[200:]

x_train = train.drop('adjclose', axis=1)
y_train = train['adjclose']
x_valid = valid.drop('adjclose', axis=1)
y_valid = valid['adjclose']

#implement linear regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```
